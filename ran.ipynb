```python
import numpy as np
import pandas as pd
```

```python
link = 'sample.xlsx'
sample = pd.read_excel(link)
sample.head()
```

```python
listOfstem = sample.to_numpy().tolist()
listOfstem
```

```python
import random
rans = random.sample(range(0, len(listOfstem)), len(listOfstem))
#len(set(rans))
```

```python
listOfrandom = [ listOfstem[a] for a in rans]
listOfrandom
```

```python
#Number of team members
N_TEAM = 4
a = 0
x = len(listOfrandom)
finalResult = []
while  x != 0:
   # print(listOfrandom[a: a+N_TEAM]) 
    if x%N_TEAM != 0 :
        print("Number of persons must be divideable on the number of team members")
        break
    finalResult.append(listOfrandom[a: a+N_TEAM]) 
    a += N_TEAM
    x -= N_TEAM
print(finalResult)
```

```python
FINAL  = pd.DataFrame(finalResult)
FINAL
#use the line below to make an excel file by the result
#FINAL.to_excel('FINAL.xlsx')
```
